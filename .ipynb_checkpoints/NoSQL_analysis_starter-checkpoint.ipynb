### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20

query = {'scores.Hygiene': 20}    
results = establishments.find(query)

first_establishment = results[0] 

for result in results:
    pprint(result)

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)

print(f"Number of establishments with a hygiene score of 20: {count}")

# Display the first document in the results using pprint
pprint(first_establishment)
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(establishments.find({'scores.Hygiene':20}))

# Display the number of rows in the DataFrame
df["_id"].count()
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
Latitude = 0.08384000
Longitude = 51.49014200

query = {
    "RatingValue": 5
}

# Execute the query and retrieve the results
results = establishments.find(query)

# Convert results to a list of dictionaries
results_list = list(results)

# Create a DataFrame from the results
df = pd.DataFrame(results_list)

# Add the new restaurant "Penang Flavours" as a reference point
new_restaurant = {
    "Name": "Penang Flavours",
    "Latitude": 0.08384000,
    "Longitude": 51.49014200
}

# Calculate the distance from each establishment to the new restaurant
df["Latitude_Diff"] = df["geocode.latitude"] - new_restaurant["Latitude"]
df["Longitude_Diff"] = df["geocode.longitude"] - new_restaurant["Longitude"]
df["Distance"] = ((df["Latitude_Diff"]) ** 2 + (df["Longitude_Diff"]) ** 2) ** 0.5

# Filter establishments within the desired range
df = df[(abs(df["Latitude_Diff"]) <= degree_search) & (abs(df["Longitude_Diff"]) <= degree_search)]

# Sort the DataFrame by hygiene score and distance
df = df.sort_values(by=["HygieneScore", "Distance"], ascending=[True, True])  

# Print the results
# Retrieve the top 5 establishments
top_5 = df.head(5)

# Print the top 5 establishments
print(top_5)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
#Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
query = {
    "scores.Hygiene": 0
}

# 2. Groups the matches by Local Authority
results = establishments.aggregate([
    {"$match": query},
    {"$group": {
        "_id": "$LocalAuthorityName",
        "count": {"$sum": 1}
    }},
    {"$sort": {"count": -1}}
])


# Print the count of establishments in each Local Authority area
for result in results:
    print(f"Local Authority: {result['_id']}, Count: {result['count']}")
    
```
